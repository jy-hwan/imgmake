```python
from fastapi import FastAPI
from pydantic import BaseModel
from fastapi.responses import StreamingResponse
from io import BytesIO
import random
from pyngrok import ngrok
import uvicorn
from threading import Thread
import requests

# === FastAPI Ïï± ===
app = FastAPI()

# === ÏÑ§Ï†ï Í∞í ===
color_palettes = ["pastel colors", "neon color palette", "vibrant rainbow colors"]
abstract_keywords = ["abstract pattern", "glowing shapes", "vaporwave style"]

# === Ïú†Ìã∏ Ìï®Ïàò ===
def get_random_palette():
    return random.choice(color_palettes)

def get_random_abstract():
    return random.choice(abstract_keywords)

# === ÏöîÏ≤≠ Ïä§ÌÇ§Îßà ===
class MoodRequest(BaseModel):
    mood: str
    color: str

# === ÎùºÏö∞Ìä∏ ===
@app.get("/")
def root():
    return {"message": "Colab FastAPI with Stable Diffusion is running!"}

@app.post("/generate")
def generate(req: MoodRequest):
    # prompt = f"K-pop album cover art, {req.mood}, {get_random_palette()}, {get_random_abstract()}, centered composition, highly detailed"

    prompt = (
    f"K-pop album cover art, themed with moods such as {req.mood}, "
    f"and colors such as {req.color}, abstract, centered composition, "
    "highly detailed, digital painting, trending on ArtStation, 4k, no people, no humans, no faces, no figures"
    )

    # Stable Diffusion ÌååÏù¥ÌîÑÎùºÏù∏ Ïã§Ìñâ (ÏòàÏãú)
    image = pipe(prompt, num_inference_steps=30).images[0]
    buf = BytesIO()
    image.save(buf, format="PNG")
    buf.seek(0)
    return StreamingResponse(buf, media_type="image/png")

# === ngrok ÌÑ∞ÎÑê Ïó¥Í∏∞ ===
public_url = ngrok.connect(8000)
print(f"üöÄ NGROK PUBLIC URL: {public_url}")

# === Ï£ºÏÜåÎ•º Ïô∏Î∂Ä Flask ÏÑúÎ≤ÑÏóê Ï†ÑÏÜ° ===
try:
    # Ïó¨Í∏∞Ïóê Flask ÏÑúÎ≤ÑÏùò ngrok Ï£ºÏÜå or Í≥†Ï†ï IP ÏÇ¨Ïö©
    flask_server_url = "https://5da2a9192109.ngrok-free.app/service/update_ngrok_url"

    res = requests.post(flask_server_url, json={"url": str(public_url)})
    if res.status_code == 200:
        print("‚úÖ Flask ÏÑúÎ≤ÑÏóê ngrok Ï£ºÏÜå Ï†ÑÏÜ° ÏÑ±Í≥µ")
    else:
        print(f"‚ö†Ô∏è Ï†ÑÏÜ° Ïã§Ìå®: ÏÉÅÌÉúÏΩîÎìú {res.status_code}")
except Exception as e:
    print(f"‚ùå Flask ÏÑúÎ≤Ñ Ï†ÑÏÜ° Ïò§Î•ò: {e}")

# === FastAPI ÏÑúÎ≤Ñ Ïã§Ìñâ ===
def run():
    uvicorn.run(app, host="0.0.0.0", port=8000)

Thread(target=run).start()
```
